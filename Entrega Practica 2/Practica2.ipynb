La inteligencia es algo que diferencia a los seres humanos de los animales, ya que a medida que fuimos evolucionando desde los neandertales, hemos adquirido diferentes conocimientos que a lo largo de los años la raza humana se ha proclamado como la más inteligente del planeta, sin embargo, a ciencia cierta ¿Qué es la inteligencia?
Lo más sencillo de decir para responder a la pregunta, es que quizás la inteligencia es aquello con lo que nos referimos a una persona o personas que tienen más conocimientos de algo, que normalmente se relacionan directamente con los conocimientos académicos, ya sea de alguien que se le dan muy bien las matemáticas y puede resolver ejercicios de manera rápida, o quizás que posea conocimientos de física o química, sin embargo, tomando eso como estándar dejamos fuera muchas otras cosas, como el conocimiento acerca de música, pintura, cuestiones físicas, etc.<br>
Y justamente de esa problemática es la que aborda La Teoría de las Inteligencias Múltiples que fue ideada por el psicólogo estadounidense Howard Gardner. Dicha teoría tiene por propósito rechazar el concepto tradicional que se tiene acerca de la inteligencia y de la manera en que esta es medida o reconocida.<br>
Debido a que Gardner trabajo en campos muy importante del área de la educación como lo son:<br>
- Los procesos de aprendizaje 
- El funcionamiento del cerebro. <br>


Según él, una vez que concluyo con sus estudios, tuvo la oportunidad de trabajar con niños que habían sufrido alguna especie de lesión cerebral, niños de diferentes culturas, talentos, etc., esto lo llevo a darse cuenta que existían niños que eran muy buenos para realizar trabajos matemáticos pero no eran buenos en otras ramas como la poesía o las letras, o casos contrarios, en lo que les apasionaba todo lo relacionado con la lectura y la escritura pero simplemente más matemáticas no se les daban. <br>

Gracias a su trabajo con los niños, Howard concluye y desarrolló su **Teoría de las inteligencias múltiples**, la cual dice *“todas las personas cuentan con distintas potencialidades, pero no se nace siendo inteligente sino que se forma gracias a la estimulación, el contexto que rodea a cada persona, cultura, tipo de educación así como el grado de motivación que tenga la persona”*. <br>
Expresado de otra forma, la inteligencia no se refiere únicamente a los conocimientos o capacidades en ramas académicas o sus derivados, sino que también cuentan como inteligencia muchas otras cosas, en su teoría, Howard logró identificar y definir la existencia de 8 tipos de inteligencia diferentes. <br>

### Los 8 tipos de inteligencia según Howard Gardner <br>

**Inteligencia lingüística** <br>
Esté tipo de inteligencia se refiere a la capacidad de comunicación, no sólo oral, sino que también engloba la comunicación a manera de escritura, expresiones corporales, entre otras formas de comunicación.<br>

Este tipo de inteligencia se encuentra presente o se puede ver reflejada con mayor facilidad en actores, poetas, escritores, etc. Quienes cuentan con capacidades de expresiones mejores que a la de la mayoria de las demás personas.<br>

**Inteligencia logico-matematica** <br>
A esta inteligencia se le consideraba la inteligencia en bruto, ya que en esta se engloban las capacidades para el razonamiento lógico y la resolución de problemas matemáticos, y esta se podría medir en la rapidez con la que se solucionan este tipo de problemas que tienen que ver con la lógica y la matemática.  <br>

Los ingenieros, matemáticos, científicos, etc., son quienes cuentan con este tipo de inteligencia, y eso puede observarse dado que en sus ramas se ven en la necesidad de resolver diferentes problemas de esta índole, que si bien con práctica o revisando el paso a paso alguien aprender o todos pueden llegar a resolver problemas matematicos, es una realidad que existen personas que tienen una mucho mayor facilidad entendimiento de ese tipo de cosas y pueden solucionar los problemas de forma mucho más rápida. 
<br>

**Inteligencia espacial** <br>
Gardner describe a este tipo de inteligencia como la habilidad que *permite observar el mundo y los objetos desde diferentes perspectivas*. <br>

Este tipo de inteligencia es caracteristico de las personas que son buenas en las artes visuales como lo son los pintores, diseñadores, escultores y demás profesionales, que usan la imaginación y la visualización de las cosas de una manera diferente. <br>
Esto se ve reflejado en museos o exposiciones de arte, en las que gente que no entiende de arte o no cuenta con este tipo de inteligencia, no ve nada más alla de simples pinturas o formas, pero por el contrario para personas que la poseen pueden comprender y ver cosas que la personas normales no. <br>

**Inteligencia musical** <br>
Según esta teoria, tomando en cuenta que alrededor el mundo todas las culturas tienen algún tipo de música, que es muy propia de ellos, es que existe una inteligencia musical. <br>
Aquí los mayores referentes con aquellos que se encuentran muy metidos en el tema de la musica, cantantes, compositores, quienes tocan algun instrumento y de más. Claro que todos podemos escuchar y disfrutar de la música, pero no todos son diestros al momento de crear música. <br>


**Inteligencia corporal y cinestésica** <br>
En palabras simples, esta inteligencia se refiere al poder utilizar tus capacidades fisicas asi como el uso de herramientas. <br>
Por razones obvias dada la definición dada, quienes destacan en este tipo de inteligencia con los deportistas, independientemente del deporte que práctiquen todos deben de utilizar sus habilidades fisicas para lograr los objetivos que necesaitan, por otro también se encuentran los cirujanos, ya que para ellos es vital el saber como utilizar las herramientas con las que trabaja. <br>

**Inteligencia intrapersonal** <br>
Hace referencia a la capacidad de control de las emociones que cada uno tiene dentro de su persona, para Gardner el desarrollo de está inteligencia hace más sencillo el sobrellevar las situaciones dificiles, ya que al poseer un mejor manejo de las emociones, ayuda a mantenerte fuerte ante las adversidades. <br>

**Inteligencia interpersonal** <br>
Esta inteligencia, nos habla acerca de la capacidad que se tiene como persona de interactuar, pero no no se refiere simplemente a un tema socializar, ya que es mucho más profundo que eso. <br>
En realidad se refiere a la habilidad de detectar y entender las circunstancias y problemas con de las demás personas, y de esta forma poder entender mejor a la otra persona, saber como se encuentra o si pasa algo. <br>
Quienes cuentan con este tipo de inteligencia, usualmente son profesores, psicólogos, terapeutas, abogados y pedagogos, que se ven muy ligados al trato con más personas en situaciones en las que las emociones con demasiado importantes. <br>

**Inteligencia naturalista**
Se dice de que es la capacidad de detectar, diferenciar y categorizar aspectos vinculados al entorno. <br>
A resumidas cuentas se le puede entender como la habilidad de reconocer las especies de animales y vegetales o fenomenos relacionados con el clima. <br>

En conclusión, la teoría de Gardber nos habla de que no hay un solo tipo de inteligencia, además de que cada persona puede contar con capacidades y habilidades diferentes. Si bien tiene muchos puntos fuertes, tambien tiene bastantes carencias, y seguún dependa la persepción de cada quien puede variar en cuanto a si el tiene razon o no.
En lo personal considero que tiene razón en que no deberia medirse la inteligencia de alguien solamente basandonos en sus conocimientos academicos, y que debemos conciderar a los que tienen habilidades musicales o de escritura, también casos como la inteligencia naturalista no las meteria como un tipo de inteligencia propio, pero bueno ya pasa a ser gusto y percepción de cada persona.
